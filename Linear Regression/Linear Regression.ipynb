```python
# Designing the single var linear regression Model

def hypothesis(theta_array, x):
  return theta_array[0] + theta_array[1] * x;
```

```python
def costfunction(theta_array, x_value, y_value, m):
  sum = 0
  for i in range(m):
    sum += ((theta_array[0] + theta_array[1] * x_value[i]) - (y_value)) ** 2
  return sum/(2 * m) 
```

```python
# Gradient Descent
def gradientDescent(theta_array,X,Y,alpha,m):
  sum_0 = 0  #theta[0]
  sum_1 = 1  #theta[1] 

  for i in range(m):
    sum_0 = sum_0 + ((theta_array[0] + theta_array[1] * X[i]) - Y[i])
    sum_1 = sum_1 + X[i] * ((theta_array[0] + theta_array[1] * X[i]) - Y[i])

  new_theta0 = theta_array[0] - alpha * (sum_0 / m)
  new_theta1 = theta_array[1] - alpha * (sum_1 / m)
  improved_thetas = [new_theta0, new_theta1]
  print(improved_thetas)
  return (improved_thetas)
```

```python
#Training Process 
def training(x_train,y_train,alpha,iters):
  m=x_train.size
  theta_0=0
  theta_1=0
  cost_values=[]
  theta_array=[theta_0,theta_1]

  for i in range(iters):
    theta_array=gradientDescent(theta_array,x_train,y_train,alpha,m)
    loss=costfunction(theta_array,x_train,y_train,m)
    cost_values.append(loss)
    ynew=theta_array[0]+theta_array[1]*x_train
    
    plt.scatter(x_train,y_train)
    plt.plot(x_train,ynew,'r')
    plt.show() 

  x=np.arange(0,iters,step=1)
  plt.plot(x,cost_values,"-g")
  plt.xlabel("number of iterations")
  plt.ylabel("Cost values")
  plt.title("Costfunction vs Iterations Curve")
  plt.show()
  return theta_array   #final theta values
```
