```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

from sklearn.datasets import make_classification
from sklearn.metrics import plot_confusion_matrix
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
X, y = make_blobs(n_samples = 100, n_features = 2, centers = 4,cluster_std = 1.5, random_state = 4) 
```

```python
plt.figure(figsize = (5, 5))
plt.scatter(X[:,0], X[:,1], c=y, marker= '*', s=100, edgecolors='red')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0) 
y_test 
```

```python
knn5 = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn5.fit(X_train, y_train)
y_pred_5 = knn5.predict(X_test)
```

```python
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
plt.scatter(X_test[:,0], X_test[:,1], c=y_pred_5, marker= '*', s=100,edgecolors='red')
plt.title("Predicted values with k=5", fontsize=20)
```

```python
# Accuracy
from sklearn.metrics import accuracy_score
print("Accuracy with k5 =>", accuracy_score(y_test, y_pred_5)*100) 
```

```python
# Confusion Matrix
print("Confusion Materix : \n")
cnf_matrix = confusion_matrix(y_test, y_pred_5, labels=[1, 0])
print(cnf_matrix)

plt.figure(figsize = (6, 5))
sn.heatmap(cnf_matrix, annot=True, cmap= 'flare',  fmt='d', cbar=True)
plt.xlabel('Predicted_Label')
plt.ylabel('Truth_Label')

plt.title('Confusion Matrix')
```
